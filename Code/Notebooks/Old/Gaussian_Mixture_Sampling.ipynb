### Gaussian Mixture Sampling

###### Omori et al (2004)

```r
library(MASS)
```

```r
sv_simul <- function(theta, log_h0, T){
    y_t = c()
    h_t = c()
    mu <- theta[1]
    phi <- theta[2]
    sigma <- theta[3]
    rho <- theta[4]
    
    mean <- matrix(c(0,0), nrow=2, ncol=1)
    std <- matrix(c(1, rho*sigma, rho*sigma, sigma**2), nrow=2, ncol=2)
    resid <- mvrnorm(T, mean, std)
    epsilon <- resid[,1]
    eta <- resid[,2]
    
    h_t[1] <- mu + phi * (h0 - mu) + eta[1]
    y_t[1] <- exp(h_t[1]/2) * epsilon[1]
    
    for (t in 2:T){
        h_t[t] <- mu + phi * (h_t[t-1] - mu) + eta[t]
        y_t[t] <- exp(h_t[t]/2) * epsilon[t]
    }
    list(y_t, h_t)
}
```

```r
# Inputs
T <- 3000
mu <- -1.2
phi <- 0.83
sigma <- 0.4
rho <- 0.5
theta <- c(mu, phi, sigma, rho) 
h0 <- 0
```

```r
data <- sv_simul(theta, log_h0, T)
y_t <- unlist(data[1])
h_t <- unlist(data[2])
```

```r
selection <- 1:300
plot(log(y_t[selection]**2), type='l')
lines(h_t[selection], type='l', col='red')
```

###### Kim (1998)

```r
sv_simul_v2 <- function(theta, T){
    y_t = c()
    h_t = c()
    mu <- theta[1]
    phi <- theta[2]
    sigma_nu <- theta[3]
    h0 <- rnorm(1, mu, sigma_nu**2/(1 -phi**2))
    
    epsilon <- rnorm(T, 0, 1)
    eta <- rnorm(T, 0, 1)
    
    h_t[1] <- mu + phi * (h0 - mu) + sigma_nu * eta[1]
    y_t[1] <- exp(h_t[1]/2) * epsilon[1]
    
    for (t in 2:T){
        h_t[t] <- mu + phi * (h_t[t-1] - mu) + sigma_nu * eta[t]
        y_t[t] <- exp(h_t[t]/2) * epsilon[t]
    }
    list(y_t, h_t)
}
```

```r
# Inputs
T <- 3000
mu <- -1.2
phi <- 0.83
sigma_nu <- 0.4
theta <- c(mu, phi, sigma_nu) 
```

```r
data <- sv_simul_v2(theta, T)
y_t <- unlist(data[1])
h_t <- unlist(data[2])
```

```r
selection <- 1:300
plot(log(y_t[selection]**2), type='l')
lines(h_t[selection], type='l', col='red')
```
